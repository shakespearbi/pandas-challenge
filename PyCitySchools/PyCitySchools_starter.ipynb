```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#total number of students
num_student = school_data_complete["Student ID"].count()
#total number of schools
num_school = school_data["school_name"].count()
#total budget
tot_budget = school_data["budget"].sum()
#average math score
avg_math = school_data_complete["math_score"].mean()
#average reading score
avg_read = school_data_complete["reading_score"].mean()
#number of students with math score 70 or greater
num_pass_math = school_data_complete.loc[school_data_complete["math_score"] >= 70]['math_score'].count()
#percentage of students with passing math score
per_pass_math = round(num_pass_math * 100/num_student,6)
#number of students with reading score 70 or greater
num_pass_read = school_data_complete.loc[school_data_complete["reading_score"] >= 70]['reading_score'].count()
#percentage of students with passing math score
per_pass_read = round(num_pass_read * 100 /num_student,6)

#passing rate
pass_rate = (avg_math + avg_read)/2
```

```python
district_dict = pd.DataFrame({"Total schools":[num_school],
                              "Total Number of Students":[num_student],
                              "Total Budget":[tot_budget],
                              "Average math score":[avg_math],
                              "Average Reading score":[avg_read],
                              "% Passing Math":[per_pass_math],
                              "% Passing Reading": [per_pass_read],
                              "%Overall Passing Rate":[pass_rate],})

#format cells
district_dict.style.format({"Total Number of Students": "{:,}",
                        "Total Budget": "${:,.2f}", 
                       "Average Reading Score": "{:.6f}", 
                       "Average Math Score": "{:.6f}", 
                       "% Passing Math": "{:.6f}", 
                       "% Passing Reading": "{:.6f}", 
                       "%Overall Passing Rate": "{:.6f}"})
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
school_grouped = school_data_complete.set_index("school_name").groupby(["school_name"])

school_type = school_data.set_index("school_name")["type"]
```

```python
count_student = school_grouped["Student ID"].count()
```

```python
budget = school_data.set_index("school_name")["budget"]
school_size = school_data.set_index("school_name")["size"]
budget_per_student = budget/school_size
```

```python
avg_math_score = school_grouped["math_score"].mean()
avg_read_score = school_grouped["reading_score"].mean()
```

```python
pass_math = school_data_complete.loc[school_data_complete["math_score"]>=70]
pass_read = school_data_complete.loc[school_data_complete["reading_score"]>=70]

per_pass_math = pass_math.groupby(["school_name"])["math_score"].count()/count_student
per_pass_reading = pass_read.groupby(["school_name"])["reading_score"].count()/count_student

overall_passrate = (per_pass_math+per_pass_reading)/2
```

```python
summary_school = pd.DataFrame({"School Type":school_type,
                               "Total Number of Students":count_student,
                              "Total Budget":budget,
                              "Budget per Student":budget_per_student,
                               "Average Math Score":avg_math_score,
                               "Average Reading Score":avg_read_score,
                               "% Passing Math":per_pass_math,
                               "% Passing Reading": per_pass_reading,
                              "Overall Passing Rate":overall_passrate})

#format cells
summary_school.style.format({"Total Budget": "${:,.2f}",
                             "Budget per Student": "${:,.2f}",
                           "Average Reading Score":"{:.6f}", 
                           "Average Math Score": "{:.6f}", 
                           "% Passing Math": "{:.6%}", 
                           "% Passing Reading": "{:.6%}", 
                           "Overall Passing Rate": "{:.6%}"})
```

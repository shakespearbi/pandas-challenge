```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#total number of students
num_student = school_data_complete["Student ID"].count()
#total number of schools
num_school = school_data["school_name"].count()
#total budget
tot_budget = school_data["budget"]
#average math score
avg_math = school_data_complete["math_score"].mean()
#average reading score
avg_read = school_data_complete["reading_score"].mean()
#passing rate
pass_rate = (avg_math + avg_read)/2
#number of students with math score 70 or greater
num_pass_math = school_data_complete.loc[school_data_complete["math_score"] >= 70]['math_score'].count()
#percentage of students with passing math score
per_pass_math = 100*num_pass_math/num_student
#number of students with reading score 70 or greater
num_pass_read = school_data_complete.loc[school_data_complete["reading_score"] >= 70]['reading_score'].count()
#percentage of students with passing math score
per_pass_read = 100*num_pass_read/num_student
```

```python
district_dict = pd.DataFrame({"Total schools":[num_school],
                              "Total Number of Students":[num_student],
                              "Total budget":[tot_budget],
                              "Average math score":[avg_math],
                              "Average reading score":[avg_read],
                              "Overall passing rate":[pass_rate],
                              "Passing Math Percentage":[per_pass_math],
                              "Passing Reading Percentage": [per_pass_read]})
district_dict
```

* Create an overview table that summarizes key metrics about each school, including:

```python
school_grouped = school_data_complete.set_index("school_name").groupby(["school_name"])
count_student = school_grouped["Student ID"].count()

budget = school_data.set_index("school_name")["budget"]

school_size = school_data.set_index("school_name")["size"]
budget_per_student = budget/school_size

school_type = school_data.set_index("school_name")["type"]

avg_math = school_grouped["math_score"].mean()
avg_read = school_grouped["reading_score"].mean()

pass_rate = (avg_math + avg_read)/2

per_pass_math = school_data_complete.loc[school_data_complete["math_score"]>=70]["math_score"].count()/count_student * 100
per_pass_read = school_data_complete.loc[school_data_complete["reading_score"]>=70]["reading_score"].count()/count_student * 100

overall = school_data_complete[(school_data_complete['reading_score'] >= 70) & (school_data_complete['math_score'] >= 70)].groupby('school_name')['Student ID'].count()/count_student 
```

```python
school_summary = pd.DataFrame({"School Type":[school_type],
                              "Total Number of Students":[count_student],
                              "Total budget":[budget_per_student],
                              "Budget per student":[tot_budget],
                              "Average math score":[avg_math],
                              "Average reading score":[avg_read],
                              "Overall passing rate":[pass_rate],
                              "Passing Math Percentage":[overall],
                              "Passing Reading Percentage": [per_pass_read]})
school_summary = school_summary[["School Type","Total Number of Students","Budget per student","Average math score",
                                "Average reading score","Overall passing rate",
                                "Passing Math Percentage","Passing Reading Percentage"]]
school_summary
```
